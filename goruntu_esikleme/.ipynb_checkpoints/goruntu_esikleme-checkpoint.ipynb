```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Resmi içe aktar
img = cv2.imread("img1.JPG")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

plt.figure()
plt.imshow(img, cmap = "gray")
plt.axis("off")
plt.show()

# Eşikleme
_, tresh_img = cv2.treshold(img, thresh = 60, maxval = 255, )
```
